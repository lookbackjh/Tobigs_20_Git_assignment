## 파생변수 생성

One_hot_Encoding

```python
object_columns = [ i for i in data.columns if i not in numeric_columns]
```

```python
new_data = data.copy()
```

```python
for i in object_columns:
    temp_data = pd.get_dummies(data[i])
    temp_data = temp_data.rename(columns=dict(zip(temp_data.columns,[i+'_'+x for x in temp_data.columns])))
    new_data = pd.concat([new_data, temp_data], axis=1)
```

```python
new_data.columns
```

나이대

```python
new_data.Age.unique()
```

```python
new_data['나이대'] = new_data.Age.apply(lambda x : '10대' if x<20 else '20대' if x<30 else '30대' if x<40 else '40대' if x<50 else '50대 이상')
```
